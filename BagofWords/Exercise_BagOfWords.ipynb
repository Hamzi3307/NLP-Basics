```python
import pandas as pd
import numpy as np
df = pd.read_csv("movies_sentiment_data.csv")
print(df.shape)
df.head()
```

```python
df = df.sample(n=19000, random_state=42)  # Use a specific random seed for reproducibility
```

#creating a new column "Category" which represent 1 if the sentiment is positive or 0 if it is negative


```python
df['Category'] = df['sentiment'].apply(lambda x: 0 if x =="negative" else 1)
```

```python
df.shape
```

#check the distribution of 'Category' and see whether the Target labels are balanced or not.


```python
df.Category.value_counts()
```

#Do the 'train-test' splitting with test size of 20%

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df["review"], df["Category"], test_size=0.30, random_state=123)
```

using sklearn pipeline module create a classification pipeline to classify the movie review's positive or negative.

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_cv = v.fit_transform(X_train)
X_test_cv = v.transform(X_test)
```

```python
X_train_cv.shape
```

```python
X_train_np = X_train_cv.toarray()
```

```python
X_train_np = X_test_cv.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```
